```python
import nltk
nltk.download('punkt')
```

```python
import os
import pandas as pd

from tqdm.auto import tqdm
from nltk.tokenize import word_tokenize
from gensim.models import Word2Vec
```

#Prepare Corpus

```python
!gdown https://drive.google.com/uc?id=11bjiwslv1gwEIWMU8YhW2k7LpVmgHmdA
!unzip /content/hotelreview.zip
```

```python
df = pd.read_csv('tripadvisor_hotel_reviews.csv')
df = df.iloc[:10000]
df
```

```python
sentences = [word_tokenize(Review.lower()) for Review in tqdm(df.Review)]
sentences[:5]
```

#Train Word2Vec Model

```python
model = Word2Vec(sentences, vector_size=128, window=5, min_count=3, workers=4, epochs=500, sg=0, hs=0)
```

#Save

```python
os.makedirs("model/w2v/", exist_ok=True)
```

```python
model.save('model/w2v/gojek_review.w2v')
```

#Load

```python
model = Word2Vec.load('model/w2v/gojek_review.w2v')
```

#Continue Training

```python
contoh_data = [
    ['it', 'it', 'a', 'great', 'day', 'in', 'here', '.', ''],
    ['what', 'a', 'piece', 'of', 'shit', '.']
]
```

```python
model.train(contoh_data, total_examples = len(contoh_data), epochs=1)
```

```python
model.save('model/w2v/gojek_review.w2v')
```
